```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
dataset = pd.read_csv("boston.csv")
```

```python
#dataset = dataset.drop(columns=['CRIM', 'AGE', 'ZN', 'INDUS', 'CHAS', 'DIS', 'RAD', 'B' ])

dataset = dataset.dropna()

cols = ['CRIM', 'TAX', 'B', 'ZN']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
nd = scaler.fit_transform(dataset[cols])
nd = pd.DataFrame(nd, columns = cols)

for col in cols:
    dataset[col] = nd[col]

dataset = dataset.dropna()
#data = data.dropna()
#sb.pairplot(dataset)
```

```python
X = dataset.drop(columns = ['MEDV', 'RAD'])

Y = dataset.iloc[:, -1]

X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1450)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(normalize=True)
regressor.fit(X_train, Y_train)
```

```python
from sklearn import metrics
pred = regressor.predict(X_test)
r_square = metrics.r2_score(Y_test, pred)
print(r_square)
```
